# Subplot

Permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder
a ellos a través de los índices.

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# Para trabajar con una fila y dos columnas se utiliza subplot
#El primer valor indica filas, segundos indica las columnas, y el tercero el índice
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y,)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'ro--')
plt.plot(y,x,'bx--')
plt.subplot(2,1,2)
plt.hist(y,)
plt.show()
```

```python
#Para dar tamaño a los plots 
z = y**2

plt.subplot(2,2,1)
plt.plot(x,y, '')
plt.subplot(1,2,2)
plt.triplot(z,x)
plt.subplot(2,2,3)
plt.triplot(z,x)
plt.show()
```

# Subplots utilizando orientados a objetos

con subplot se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices.

```python
#Una forma rápida de hacer una figura
x = np.linspace(0,5,20)
y = np.sin(x)
fig , axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
# Dado que subplot trabaja los axes como si estos fueran arrays
# Pueden ser entonces trabajados dentro de una tupla

fig, (axes1, axes2) = plt.subplots(nrows=1,ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x))
axes[0,2].plot(x,np.tan(x))
axes[0,3].plot(x,np.cos(x)**2)
axes[1,0].plot(x,np.cos(x)**2+np.sin(x)**2)
axes[1,1].plot(x,np.exp(x))
axes[1,2].plot(x,1+x+x*+2/2+x**3/6)
axes[1,3].plot(x,np.log(x+1))
#Para que las imágenes no se vean tan juntas
fig.tight_layout()
```

# Leyendas, etiquetas, títulos, tamaño

Para dar contexto a nuestros gráficos necesitamos usar títuos leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto m

```python
x = np.linspace(0,5,11)
y = np.sin(x)

fig, axes = plt.subplots(1,2, figsize = (10,5))
axes[0].plot(x,y, label = 'sin(x)')
axes[0].set_title('Relación X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('sin(x)')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('sin(x)')
axes[1].set_ylabel('X')
```

```python
# Con loc podemos cambiar la ubicación de la legenda

plt.plot(x,y, label='sin(x)')
plt.title('Este es un título')
plt.legend(loc='lower left', bbox_to_anchor=(0.3,0.3))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use("dark_background")

for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # very light grey

for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#212946'  # bluish dark grey

colors = [
    '#08F7FE',  # teal/cyan
    '#FE53BB',  # pink
    '#F5D300',  # yellow
    '#00ff41',  # matrix green
]


df = pd.DataFrame({'A': [1, 3, 9, 5, 2, 1, 1],
                   'B': [4, 5, 5, 7, 9, 8, 6]})

fig, ax = plt.subplots()

df.plot(marker='o', color=colors, ax=ax)

# Redraw the data with low alpha and slighty increased linewidth:
n_shades = 10
diff_linewidth = 1.05
alpha_value = 0.3 / n_shades

for n in range(1, n_shades+1):

    df.plot(marker='o',
            linewidth=2+(diff_linewidth*n),
            alpha=alpha_value,
            legend=False,
            ax=ax,
            color=colors)

# Color the areas below the lines:
for column, color in zip(df, colors):
    ax.fill_between(x=df.index,
                    y1=df[column].values,
                    y2=[0] * len(df),
                    color=color,
                    alpha=0.1)

ax.grid(color='#2A3459')

ax.set_xlim([ax.get_xlim()[0] - 0.2, ax.get_xlim()[1] + 0.2])  # to not have the markers cut off
ax.set_ylim(0)

plt.show()
```

# Seaborn

Seaborn es una librería construida sobre Matplotlib, por lo que hereda todas las bondades de la misma. Fue escrita por Michael Waskom y está integrada para estructuras de Pandas por lo que está optimizada para funcionar con DataFrames.

Seaborn tiene diferentes ventajas y entre ellas encuentras principalmente:

- Tiene una gran velocidad
- Facilidad para escribir código
- Altamente customizable entre gráficas y visualizaciones

## Tipos de Gráficas que tiene Seaborn
Seaborn ofrece ciertas características principales para problemas específicos de visualización de datos:

- Diagramas o gráficas relacionables
- Distribución de datos
- Graficar variables categóricas

Por ejemplo:

- Relplot (relacional): scatterplot, lineplot.
- Displot (distribuciones): histplot, kdeplot, ecdfplot, rugplot.
- Catplot (categorica): stripplot, swamplot, boxplot, violinplot, pointplot, barplot.
